```python
#Load JSON file

#printing ranges (NumPy)

import numpy as np
import pandas as pd
import datetime
```

```python
filename = "transaction-data-adhoc-analysis.json"

df = pd.read_json(filename)

print(df)
df.head
df.drop(labels=["address","birthdate","mail","transaction_items", "sex", "username","transaction_value"], axis=1)
```

```python
new_df=df[["name","transaction_date"]].copy()
new_df
new_df["month"]=pd.DatetimeIndex(new_df["transaction_date"]).month
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (First Month)
def check (x):
    if x==1:
        return True
    else:
        return False

new_df["First Month"]=new_df["month"].apply(check)
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (Second Month)
def check (x):
    if x==2:
        return True
    else:
        return False

new_df["Second Month"]=new_df["month"].apply(check)
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (Third Month)
def check (x):
    if x==3:
        return True
    else:
        return False

new_df["Third Month"]=new_df["month"].apply(check)
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (Fourth Month)
def check (x):
    if x==4:
        return True
    else:
        return False

new_df["Fourth Month"]=new_df["month"].apply(check)
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (Second Month)
def check (x):
    if x==5:
        return True
    else:
        return False

new_df["Fifth Month"]=new_df["month"].apply(check)
new_df
```

```python
#Apply Function for the Checking of the Months Per Value (Sixth Month)
def check (x):
    if x==6:
        return True
    else:
        return False

new_df["Sixth Month"]=new_df["month"].apply(check)
new_df.head(20)
```

```python
#Determine if customer is repeater

final_month1_table3['Month_2 Repeaters'] = np.where((final_month1_table3["month 1"] > 0) & (final_month1_table3["month 2"] >0), True, False)

final_month1_table3['Month_3 Repeaters'] = np.where((final_month1_table3["month 2"] > 0) & (final_month1_table3["month 3"] >0), True, False)

final_month1_table3['Month_4 Repeaters'] = np.where((final_month1_table3["month 3"] > 0) & (final_month1_table3["month 4"] >0), True, False)

final_month1_table3['Month_5 Repeaters'] = np.where((final_month1_table3["month 4"] > 0) & (final_month1_table3["month 5"] >0), True, False)

final_month1_table3['Month_6 Repeaters'] = np.where((final_month1_table3["month 5"] > 0) & (final_month1_table3["month 6"] >0), True, False)
final_month1_table3
```

```python
#Determine the sum if repeater customers per month
repeaters_month1=0
repeaters_month2=final_month1_table3['Month_2 Repeaters'].sum()
repeaters_month2
```

```python
repeaters_month3=final_month1_table3['Month_3 Repeaters'].sum()
repeaters_month4=final_month1_table3['Month_4 Repeaters'].sum()
repeaters_month5=final_month1_table3['Month_5 Repeaters'].sum()
repeaters_month6=final_month1_table3['Month_6 Repeaters'].sum()
```

```python
#Determine if customer is inactive
inactive_table=final_month1_table3[["name", "month 1", "month 2", "month 3", "month 4", "month 5", "month 6"]].copy()
inactive_table
```

```python
#Determine if customer is inactive
inactive_table['Month_2 Inactive'] = np.where((inactive_table["month 1"] > 0) & (inactive_table["month 2"]==0), True, False)
inactive_table['Month_3 Inactive'] = np.where(((inactive_table["month 1"] > 0) | (inactive_table["month 2"]>0))&(inactive_table["month 3"]==0), True, False)
inactive_table['Month_4 Inactive'] = np.where(((inactive_table["month 1"] > 0) | (inactive_table["month 2"]>0) | (inactive_table["month 3"]>0))&(inactive_table["month 4"]==0), True, False)
inactive_table['Month_5 Inactive'] = np.where(((inactive_table["month 1"] > 0) | (inactive_table["month 2"]>0) | (inactive_table["month 3"]>0) | (inactive_table["month 4"]>0))&(inactive_table["month 5"]==0), True, False)
inactive_table['Month_6 Inactive'] = np.where(((inactive_table["month 1"] > 0) | (inactive_table["month 2"]>0) | (inactive_table["month 3"]>0) | (inactive_table["month 4"]>0) | (inactive_table["month 5"]>0))&(inactive_table["month 6"]==0), True, False)
```

```python
#Determine sum of inactive customers per month
inactive_count_month1=0
inactive_count_month2=inactive_table['Month_2 Inactive'].sum()
inactive_count_month3=inactive_table['Month_3 Inactive'].sum()
inactive_count_month4=inactive_table['Month_4 Inactive'].sum()
inactive_count_month5=inactive_table['Month_5 Inactive'].sum()
inactive_count_month6=inactive_table['Month_6 Inactive'].sum()
```

```python
#Determine if customer is engaged
engaged_table=final_month1_table3[["name", "month 1", "month 2", "month 3", "month 4", "month 5", "month 6"]].copy()
engaged_table
```

```python
#Determine if customer is engaged
engaged_table['Month_1 Engaged'] = np.where((engaged_table["month 1"] > 0), True, False)
engaged_table['Month_2 Engaged'] = np.where((engaged_table["month 1"] > 0)&(engaged_table["month 2"] > 0), True, False)
engaged_table['Month_3 Engaged'] = np.where((engaged_table["month 1"] > 0)&(engaged_table["month 2"] > 0)&(engaged_table["month 3"] > 0), True, False)
engaged_table['Month_4 Engaged'] = np.where((engaged_table["month 1"] > 0)&(engaged_table["month 2"] > 0)&(engaged_table["month 3"] > 0)&(engaged_table["month 4"] > 0), True, False)
engaged_table['Month_5 Engaged'] = np.where((engaged_table["month 1"] > 0)&(engaged_table["month 2"] > 0)&(engaged_table["month 3"] > 0)&(engaged_table["month 4"] > 0)&(engaged_table["month 5"] > 0), True, False)
engaged_table['Month_6 Engaged'] = np.where((engaged_table["month 1"] > 0)&(engaged_table["month 2"] > 0)&(engaged_table["month 3"] > 0)&(engaged_table["month 4"] > 0)&(engaged_table["month 5"] > 0)&(engaged_table["month 6"] > 0), True, False)
engaged_table
```

```python
#Determine sum of engaged

engaged_count_month1=engaged_table['Month_1 Engaged'].sum()
engaged_count_month2=engaged_table['Month_2 Engaged'].sum()
engaged_count_month3=engaged_table['Month_3 Engaged'].sum()
engaged_count_month4=engaged_table['Month_4 Engaged'].sum()
engaged_count_month5=engaged_table['Month_5 Engaged'].sum()
engaged_count_month6=engaged_table['Month_6 Engaged'].sum()
```

```python
#Make Dataframe with the sum of the repeaters, engaged, and inactive
Data_of_Customers= {'Types of Customers': ['Repeaters', 'Inactive', 'Engaged'],'1 (January)': [repeaters_month1, inactive_count_month1, engaged_count_month1],"2 (February)":[repeaters_month2, inactive_count_month2, engaged_count_month2],"3 (March)":[repeaters_month3, inactive_count_month3, engaged_count_month3],"4 (April)":[repeaters_month4, inactive_count_month4,  engaged_count_month4],"5 (May)":[repeaters_month5, inactive_count_month5, engaged_count_month5],"6 (June)":[repeaters_month6, inactive_count_month6, engaged_count_month6]}
CustomerTypte_df = pd.DataFrame(Data_of_Customers).set_index('Types of Customers')
```
